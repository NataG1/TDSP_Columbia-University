**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Colab Notebooks/data/nyc_collisions.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour.values, palette="viridis")
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Crashes', fontsize=14)
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonal Component (Yearly Patterns)', fontsize=14)
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "orange"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "gray" # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color= color,
          fill=True,
          fill_color= color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
# Left join to retain all crashes, even if person data is missing
merged_df = pd.merge(
    crash_df,
    drivers_df,
    on='COLLISION_ID',
    how='left'
)
````

````python
# explore data
merged_df.info()
merged_df.describe()
merged_df.head()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Prepare the data
analysis_df = merged_df[
    (merged_df['DRIVER_SEX'].isin(['M', 'F'])) &
    (merged_df['CONTRIBUTING_FACTOR_1'].notna())
].copy()

factor_mapping = {
    'unsafe speed': 'speeding',
    'speed related': 'speeding',
    'alcohol involvement': 'dui',
    'drugs (illegal)': 'dui',
    'intoxicated': 'dui',
    'driver inattention/distraction': 'distraction',
    'distraction/inattention': 'distraction',
    'cell phone (hand-held)': 'distraction',
    'texting': 'distraction',
    'aggressive driving/road rage': 'aggressive',
    'driver inexperience': 'inexperience',
    'pavement slippery': 'road condition',
    'unspecified': 'other',
    'none': 'other'
}

analysis_df['MAIN_FACTOR'] = (
    analysis_df['CONTRIBUTING_FACTOR_1']
    .str.strip().str.lower()
    .replace(factor_mapping)
)

# 2. Calculate percentages within each gender
top_factors = (
    analysis_df[analysis_df['MAIN_FACTOR'] != 'other']
    ['MAIN_FACTOR'].value_counts()
    .nlargest(10).index
)

# Get counts by gender and factor
counts_df = (
    analysis_df[analysis_df['MAIN_FACTOR'].isin(top_factors)]
    .groupby(['DRIVER_SEX', 'MAIN_FACTOR'])
    .size()
    .unstack()
)

# Calculate percentages within each gender
percent_df = counts_df.apply(lambda x: x/x.sum()*100, axis=1).stack().reset_index(name='percentage')

# 3. Create separate plots for each gender
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle('Collision Cause Distribution by Gender', fontsize=16)

# Male drivers plot
sns.barplot(
    data=percent_df[percent_df['DRIVER_SEX'] == 'M'],
    y='MAIN_FACTOR',
    x='percentage',
    order=top_factors,
    color='blue',
    ax=ax1
)
ax1.set_title('Male Drivers')
ax1.set_xlabel('Percentage of Male Collisions')
ax1.set_ylabel('Contributing Factor')
ax1.set_xlim(0, percent_df['percentage'].max()*1.1)

# Female drivers plot
sns.barplot(
    data=percent_df[percent_df['DRIVER_SEX'] == 'F'],
    y='MAIN_FACTOR',
    x='percentage',
    order=top_factors,
    color='red',
    ax=ax2
)
ax2.set_title('Female Drivers')
ax2.set_xlabel('Percentage of Female Collisions')
ax2.set_ylabel('')
ax2.set_xlim(0, percent_df['percentage'].max()*1.1)

# Add percentage labels
for ax in (ax1, ax2):
    for p in ax.patches:
        width = p.get_width()
        if width > 0:
            ax.text(width + 0.5,
                    p.get_y() + p.get_height()/2,
                    f'{width:.1f}%',
                    va='center')

plt.tight_layout()
plt.show()

# 4. Print the percentage table
print("Percentage distribution within each gender group:")
display(
    percent_df.pivot(index='MAIN_FACTOR', columns='DRIVER_SEX', values='percentage')
    .round(1)
    .sort_values('M', ascending=False)
)
````

````python
#Statistical significance testing (chi-square)

from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(
    analysis_df['DRIVER_SEX'],
    analysis_df['MAIN_FACTOR']
)

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square p-value: {p:.4f}")
print("Expected frequencies if no association:")
print(pd.DataFrame(expected, columns=contingency_table.columns,
                  index=contingency_table.index).round(1))
````

````python
import numpy as np

# Calculate Cramer's V (measure of association strength)
n = contingency_table.sum().sum()
phi = np.sqrt(chi2/n)
r, c = contingency_table.shape
cramers_v = phi / np.sqrt(min(r-1,c-1))
print(f"Cramer's V: {cramers_v:.3f}")

# 2. Calculate standardized residuals
residuals = (contingency_table - expected) / np.sqrt(expected)

# 3. Plot the most significant deviations
plt.figure(figsize=(14, 6))
top_factors = 10  # Show top 10 most divergent factors

# Get factors with largest absolute residuals
significant_factors = residuals.abs().max().sort_values(ascending=False).index[:top_factors]

sns.heatmap(
    residuals[significant_factors].T,
    annot=True,
    fmt=".1f",
    cmap='coolwarm',
    center=0,
    cbar_kws={'label': 'Standardized Residual'}
)
plt.title('Gender Differences in Collision Causes\n(Observed - Expected, in standard deviations)')
plt.tight_layout()
plt.show()

````

While the chi-square test was statistically significant (p < 0.0001), this does not imply meaningful real-world differences.

The tiny effect size suggests gender explains less than 1% of the variance in collision causes.

Observed differences (e.g., males 6.8% vs females 5.8% in "passing/lane usage") are statistically detectable but practically minor.

````python
# Calculate relative risk (Male/Female)
relative_risk = (counts_df.loc['M']/counts_df.loc['M'].sum()) / \
                (counts_df.loc['F']/counts_df.loc['F'].sum())

plt.figure(figsize=(10,6))
sns.barplot(
    x=relative_risk.sort_values(ascending=False).values,
    y=relative_risk.sort_values(ascending=False).index,
    palette='viridis'
)
plt.axvline(1, color='red', linestyle='--')
plt.title('Relative Risk of Collision Causes (Male/Female)')
plt.xlabel('Relative Risk Ratio')
plt.ylabel('Collision Cause')
plt.tight_layout()
plt.show()
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report

# Prepare data for modeling
model_df = analysis_df[analysis_df['MAIN_FACTOR'].isin(key_factors)].copy()
X = pd.get_dummies(model_df['MAIN_FACTOR'])
y = model_df['DRIVER_SEX'].map({'M':1, 'F':0})

# Train model
model = LogisticRegression()
model.fit(X, y)

# Interpret results
coef_df = pd.DataFrame({
    'factor': X.columns,
    'coef': model.coef_[0],
    'odds_ratio': np.exp(model.coef_[0])
}).sort_values('odds_ratio', ascending=False)

print("Factors predicting male vs female drivers:")
display(coef_df)
````
